```python
import numpy as np
import pandas as pd

data = pd.read_csv('war12.csv', encoding='utf-8-sig', sep=';')

data = data.applymap(lambda x: str(x).replace(',', '.') if isinstance(x, str) else x)

data = data.astype(float)

X = data[['x1', 'x2']].values
Y = data['y'].values.reshape(-1, 1)

X_pseudo_inv = np.linalg.pinv(X)

beta = X_pseudo_inv @ Y
a, b = beta.flatten()

print(f'a = {a}')
print(f'b = {b}')
```
